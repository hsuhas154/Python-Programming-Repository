```python
import time
from IPython.display import clear_output
a, b, c, d, e, f, g, h, i = ' ',' ',' ',' ',' ',' ',' ',' ',' '
char1 = ''
char2 = ''

## FUNCTION SEPARATING COMMENT ##

def game_instructions():
        
    flag = True
    
    while flag:
        print('Here is the game explanation:')
        print('This game can be played by two players only.')
        print('Each player has to choose their character. Either X or O.')
        print('Player 1 goes first and Player 2 goes next.')
        print('Each player marks Xs and Os on the grid alternatively one after the other.')
        print('If any player successfully manages to enter all Xs or Os in one straight line first, either horizontally, vertically or diagonally, that person wins the game.')
        print('Each cell in the grid is identified by its cell number.')
        print('The grid and the cell numbers will be shown below.')
        print('Note that each player has to specify the corresponding cell number in which they wish to mark either X or O.')
        print('\n')
        print('Have you understood the instructions so far?')
        print('Press 1 for YES and 0 to read all instructions again.')
        user_key = input('Enter your response: \n')
        
        if user_key == '1':
            flag = False
            clear_output()
        elif user_key == '0':
            clear_output()
        else:
            print('Please enter a valid response. You will be asked to enter your response again.')
            time.sleep(4)
            clear_output()
    
    return

## FUNCTION SEPARATING COMMENT ##

def startup_messages():
    print('WELCOME TO TIC TAC TOE')
    time.sleep(5)
    flag = True
        
    while flag:
        user_input = input('PRESS 1 TO CONTINUE\n')
        
        if user_input == '1':
            print('PLEASE WAIT')
            time.sleep(4)
            clear_output()
            flag = False
        else:
            print('ENTERING ANYTHING OTHER THAN 1 IS NOT ALLOWED. YOU WILL BE ASKED TO ENTER YOUR RESPONSE AGAIN.')
        
    return

## FUNCTION SEPARATING COMMENT ##

def game_layout():
    print('Below diagram depicts the grid and the cell numbers of each cell associated with the grid.')
    print('This diagram will be made available throughout the gameplay for the convenience of the gamers.')
    print('Player 1 will go first and then Player 2 will follow!\n')
    
    print('       |       |       ')
    print('   7   |   8   |   9   ')
    print('_______|_______|_______')
    print('       |       |       ')
    print('   4   |   5   |   6   ')
    print('_______|_______|_______')
    print('       |       |       ')
    print('   1   |   2   |   3   ')
    print('       |       |       ')
    
## FUNCTION SEPARATING COMMENT ##    
    
def character_decision():
    global char1
    global char2
    flag1 = True
    
    while flag1:
        char1 = input('Player 1, Choose a character (X or O). Please enter only X or O: ')
        
        if char1 == 'X':
            print('\nPlayer 2 will get the character O')
            char2 = 'O'
            flag1 = False
        elif char1 == 'O':
            print('\nPlayer 2 will get the character X')
            char2 = 'X'
            flag1 = False
        else:
            print('Invalid Input. You have to enter only an X or an O. You will be asked to enter your response again.')
            time.sleep(4)
    
    time.sleep(6)
    return

## FUNCTION SEPARATING COMMENT ##

def game_play_1(x):
    
    global a
    global b
    global c
    global d
    global e
    global f
    global g
    global h
    global i
    global char1
    global char2
    
    if x == 1:
        a = char1
    elif x == 2:
        b = char1
    elif x == 3:
        c = char1
    elif x == 4:
        d = char1
    elif x == 5:
        e = char1
    elif x == 6:
        f = char1
    elif x == 7:
        g = char1
    elif x == 8:
        h = char1
    elif x == 9:
        i = char1
        
    print(f'         |         |         ')
    print(f'     {g}   |    {h}    |   {i}   ')
    print(f'_________|_________|_________')
    print(f'         |         |         ')
    print(f'     {d}   |    {e}    |   {f}   ')
    print(f'_________|_________|_________')
    print(f'         |         |         ')
    print(f'     {a}   |    {b}    |   {c}   ')
    print(f'         |         |         ')
    
## FUNCTION SEPARATING COMMENT ##

def game_play_2(y):
    
    global a
    global b
    global c
    global d
    global e
    global f
    global g
    global h
    global i
    global char1
    global char2
    
    if y == 1:
        a = char2
    elif y == 2:
        b = char2
    elif y == 3:
        c = char2
    elif y == 4:
        d = char2
    elif y == 5:
        e = char2
    elif y == 6:
        f = char2
    elif y == 7:
        g = char2
    elif y == 8:
        h = char2
    elif y == 9:
        i = char2
        
    print(f'         |         |         ')
    print(f'     {g}   |    {h}    |   {i}   ')
    print(f'_________|_________|_________')
    print(f'         |         |         ')
    print(f'     {d}   |    {e}    |   {f}   ')
    print(f'_________|_________|_________')
    print(f'         |         |         ')
    print(f'     {a}   |    {b}    |   {c}   ')
    print(f'         |         |         ')

## FUNCTION SEPARATING COMMENT ##

def game_input():
    
    global a
    global b
    global c
    global d
    global e
    global f
    global g
    global h
    global i
    global char1
    global char2
    
    a, b, c, d, e, f, g, h, i = ' ',' ',' ',' ',' ',' ',' ',' ',' '
    x, y = 0, 0
    t = 1
    
    print('\n')   
    print('         |       |       ')
    print('     7   |   8   |   9   ')
    print('  _______|_______|_______')
    print('         |       |       ')
    print('     4   |   5   |   6   ')
    print('  _______|_______|_______')
    print('         |       |       ')
    print('     1   |   2   |   3   ')
    print('         |       |       ')   
    print('\n')
        
    while t <= 5:

        if t == 1:
            x = int(input('Player 1\'s move: '))
        else:
            check_x = True
            while check_x:
                x = int(input('Player 1\'s move: '))
                
                if x == y:
                    print('You have entered the same cell number as that of Player 2. This is not allowed')
                    print('You will be asked to enter a different cell number again.')
                else:    
                    check_x = False
        
        clear_output()
        print('\n')
        print('         |       |       ')
        print('     7   |   8   |   9   ')
        print('  _______|_______|_______')
        print('         |       |       ')
        print('     4   |   5   |   6   ')
        print('  _______|_______|_______')
        print('         |       |       ')
        print('     1   |   2   |   3   ')
        print('         |       |       ')
        print('\n')
        game_play_1(x)
        
        if (((a == char1) and (b == char1) and (c == char1))
            or
            ((d == char1) and (e == char1) and (f == char1))
            or
            ((g == char1) and (h == char1) and (i == char1))
            or
            ((a == char1) and (d == char1) and (g == char1))
            or
            ((b == char1) and (e == char1) and (h == char1))
            or
            ((c == char1) and (f == char1) and (i == char1))
            or
            ((a == char1) and (e == char1) and (i == char1))
            or
            ((c == char1) and (e == char1) and (g == char1))):
                print('\nPLAYER 1 HAS WON THE GAME!\n')
                x = 0
                y = 0
                return
            
        if t == 5:
            print('\nTHE GAME IS A DRAW\n')
            x = 0
            y = 0
            return
        
        check_y = True
        while check_y:
            y = int(input('Player 2\'s move: '))
            
            if y == x:
                print('You have entered the same cell number as that of Player 1. This is not allowed')
                print('You will be asked to enter a different cell number again.')
            else:    
                check_y = False
                
        clear_output()
        print('\n')
        print('         |       |       ')
        print('     7   |   8   |   9   ')
        print('  _______|_______|_______')
        print('         |       |       ')
        print('     4   |   5   |   6   ')
        print('  _______|_______|_______')
        print('         |       |       ')
        print('     1   |   2   |   3   ')
        print('         |       |       ')
        print('\n')
        game_play_2(y)
        
        if (((a == char2) and (b == char2) and (c == char2))
            or
            ((d == char2) and (e == char2) and (f == char2))
            or
            ((g == char2) and (h == char2) and (i == char2))
            or
            ((a == char2) and (d == char2) and (g == char2))
            or
            ((b == char2) and (e == char2) and (h == char2))
            or
            ((c == char2) and (f == char2) and (i == char2))
            or
            ((a == char2) and (e == char2) and (i == char2))
            or
            ((c == char2) and (e == char2) and (g == char2))):
                print('\nPLAYER 2 HAS WON THE GAME!\n')
                x = 0
                y = 0
                return
        
        t = t + 1
    
    x = 0
    y = 0
    return

# FUNCTION SEPARATING COMMENT

def game_outro():
    clear_output()
    print('THANK YOU FOR PLAYING TIC TAC TOE')
    
# FUNCTION SEPARATING COMMENT
```
